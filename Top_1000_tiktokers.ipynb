```python
# finding empty cells
df.isna().sum()
```

```python
# filling those empty cells 
df = df.dropna(axis=0)
df.isna().sum()
```

```python
# Convert K and M to numbers
def symbol(x):
    if type(x) == float or type(x) == int:
        return x
    if 'M' in x:
        if len(x)>1:
            return float(x.replace('M',''))*1000000
        return 1000000.0
    if 'K' in x:
        if len(x)>1:
            return float(x.replace('K', ''))*1000
        return 1000.0
```

```python
df['Subscribers Count'] = df['Subscribers Count'].apply(symbol)
df['Views. Avg'] = df['Views. Avg'].apply(symbol)
df['Likes. Avg'] = df['Likes. Avg'].apply(symbol)
df['Comments. Avg'] = df['Comments. Avg'].apply(symbol)
df['Shares. Avg'] = df['Shares. Avg'].apply(symbol)
df.head()
```

```python
# Total Interaction with each tiktoker
index_likes = df.columns.get_loc('Likes. Avg')
index_shares = df.columns.get_loc('Shares. Avg')

df['Total Interaction'] = df.iloc[:,index_likes:index_shares+1].sum(axis=1)
df.sort_values(['Total Interaction'], ascending = False)
```

```python
# most liked tiktokers
df.sort_values(['Likes. Avg'], ascending = False)
```

```python
# most subs tiktokers
df.sort_values(['Subscribers Count'], ascending = False)
```
